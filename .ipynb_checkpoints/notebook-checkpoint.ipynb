### Cleaning Data

#### Order Dataset

```python
# orders_df
```

```python
orders_df.info()
```

```python
datetime_columns = ["order_approved_at","order_delivered_carrier_date",'order_delivered_customer_date','order_estimated_delivery_date']
 
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
  orders_df[column] = orders_df[column].dt.date
  orders_df[column] = pd.to_datetime(orders_df[column])
orders_df.info()
```

```python
datetime_columns = ["order_purchase_timestamp"]
 
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
  orders_df[column] = orders_df[column].dt.strftime("%H")
orders_df.head()
```

```python
orders_df[orders_df.order_approved_at.isna()]
```

```python
orders_df = orders_df.sort_values('order_approved_at')
```

```python
orders_df['order_approved_at'] = orders_df['order_approved_at'].ffill()
```

```python
orders_df = orders_df.sort_values('order_delivered_carrier_date')
```

```python
orders_df['order_delivered_carrier_date'] = orders_df['order_delivered_carrier_date'].ffill()
```

```python
orders_df = orders_df.sort_values('order_delivered_customer_date')
```

```python
orders_df['order_delivered_customer_date'] = orders_df['order_delivered_customer_date'].ffill()
```

```python
orders_df.isna().sum()
```

**Insight:**
- Terlihat bahwa missing value yang dimiliki ketiga feature order_approved_at, order_delivered_carrier_date, order_delivered_customer_date selalu berpasangan.
- kemudian menggunakan method ffil untuk mengisi nilai yang kosong dengan nilai sebelumnya.

```python
# order_reviews_df
```

```python
order_reviews_df[order_reviews_df.review_comment_title.isna()]
```

```python
order_reviews_df[order_reviews_df.review_comment_message.isna()]
```

```python
order_reviews_df_clean = order_reviews_df.drop(['review_comment_title', 'review_comment_message'], axis=1)
```

```python
datetime_columns = ["review_creation_date"]
 
for column in datetime_columns:
  order_reviews_df_clean[column] = pd.to_datetime(order_reviews_df_clean[column])
  order_reviews_df_clean[column] = order_reviews_df_clean[column].dt.date
  order_reviews_df_clean[column] = pd.to_datetime(order_reviews_df_clean[column])
order_reviews_df_clean.info()
```

```python
datetime_columns = ["review_answer_timestamp"]

for column in datetime_columns:
  order_reviews_df_clean[column] = pd.to_datetime(order_reviews_df_clean[column])
  order_reviews_df_clean[column] = order_reviews_df_clean[column].dt.strftime("%H")
order_reviews_df_clean.head(5)
```

**Insight:**
- Karena terdapat sangat bayak missing value, maka saya melakukan drop pada kolom 'review_comment_title', 'review_comment_message'.

```python
new_orders_df['order_year'] = new_orders_df['order_approved_at'].dt.year
new_orders_df['order_month'] = new_orders_df['order_approved_at'].dt.month
```

```python
# membuat feature order_year dan order_month berdasarkan order_approved_at
def get_month_name(month):
    return (
        "january" if month == 1
        else
        "february" if month == 2
        else
        "march" if month == 3
        else
        "april" if month == 4
        else
        "may" if month == 5
        else
        "june" if month == 6
        else
        "july" if month == 7
        else
        "august" if month == 8
        else
        "september" if month == 9
        else
        "october" if month == 10
        else
        "november" if month == 11
        else
        "desember"
    )

new_orders_df['order_month_name'] = new_orders_df['order_month'].astype('int').map(get_month_name)
new_orders_df[['order_year','order_month_name']].head()
```

```python
print(new_orders_df.order_month_name.unique())
```

```python
print(new_orders_df.order_year.unique())
```

```python
new_orders_df.groupby(by='order_month_name').agg({
    "customer_id": "nunique"
}).sort_values(by="customer_id",ascending=False).reset_index()
```

```python
# Memberi label waktu transaksi pelanggan

def get_part_of_day(hour):
    return (
        "morning" if 5 <= hour <= 11
        else
        "afternoon" if 12 <= hour <= 17
        else
        "evening" if 18 <= hour <= 22
        else
        "night"
    )

new_orders_df['order_purchase_time'] = new_orders_df['order_purchase_timestamp'].astype('int').map(get_part_of_day)
new_orders_df['review_answer_time'] = new_orders_df["review_answer_timestamp"].astype('int').map(get_part_of_day)

new_orders_df[['order_purchase_time','review_answer_time']].head()
```

```python
new_orders_df.groupby(by="order_purchase_time").agg({
    "customer_id": "nunique"
}).sort_values(by="customer_id",ascending=False).reset_index()
```

```python
new_orders_df.groupby(by='review_answer_time').agg({
    "customer_id": "nunique"
}).sort_values(by="customer_id",ascending=False).reset_index()
```

### Pertanyaan 1: Bagaimana tingkat penjualan perusahaan dalam beberapa tahun terakhir?

```python
df_tingkat_penjualan_2016 = all_df[all_df.order_year == 2016]

tingkat_penjualan_2016 = df_tingkat_penjualan_2016.groupby(by='order_month').agg({
    "order_id": "nunique",
    "order_month_name" : pd.Series.mode,
    "payment_value" : "sum"
}).sort_values(by="order_month").reset_index()

x = tingkat_penjualan_2016["order_month_name"]
y = tingkat_penjualan_2016["order_id"]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data with lines and markers
ax.plot(x, y, marker='o', color='red', markersize=8, linewidth=3)

# Annotate each point with the order ID
for i, txt in enumerate(y):
    ax.annotate(str(txt), (x[i], y[i] + 15), fontsize=12, ha='center')

# Add titles and labels
ax.set_title("Tingkat Penjualan Tahun 2016", fontsize=16)
ax.set_xlabel("Order Month", fontsize=14)
ax.set_ylabel("Number of orders", fontsize=14)

# Customize the tick labels
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
```

```python
df_tingkat_penjualan_2017 = all_df[all_df.order_year == 2017]

tingkat_penjualan_2017 = df_tingkat_penjualan_2017.groupby(by='order_month').agg({
    "order_id": "nunique",
    "order_month_name" : pd.Series.mode,
    "payment_value" : "sum"
}).sort_values(by="order_month").reset_index()

x = tingkat_penjualan_2017["order_month_name"]
y = tingkat_penjualan_2017["order_id"]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data with lines and markers
ax.plot(x, y, marker='o', color='red', markersize=8, linewidth=2)

# Annotate each point with the order ID
for i, txt in enumerate(y):
    ax.annotate(str(txt), (x[i], y[i] + 15), fontsize=12, ha='center')

# Add titles and labels
ax.set_title("Tingkat Penjualan Tahun 2017", fontsize=16)
ax.set_xlabel("Order Month", fontsize=14)
ax.set_ylabel("Number of orders", fontsize=14)

# # Customize the tick labels
ax.tick_params(axis='x', rotation=45)
```

```python
df_tingkat_penjualan_2018 = all_df[all_df.order_year == 2018]

tingkat_penjualan_2018 = df_tingkat_penjualan_2018.groupby(by='order_month').agg({
    "order_id": "nunique",
    "order_month_name" : pd.Series.mode,
    "payment_value" : "sum"
}).sort_values(by="order_month").reset_index()

x = tingkat_penjualan_2018["order_month_name"]
y = tingkat_penjualan_2018["order_id"]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data with lines and markers
ax.plot(x, y, marker='o', color='red', markersize=8, linewidth=2)

# Annotate each point with the order ID
for i, txt in enumerate(y):
    ax.annotate(str(txt), (x[i], y[i] + 15), fontsize=12, ha='center')

# Add titles and labels
ax.set_title("Tingkat Penjualan Tahun 2018", fontsize=16)
ax.set_xlabel("Order Month", fontsize=14)
ax.set_ylabel("Number of orders", fontsize=14)

# # Customize the tick labels
ax.tick_params(axis='x', rotation=45)
```

### Pertanyaan 2: Bagaimana pesebaran daerah tempat pelanggan yang melakukan transaksi ?

```python
bystate_df = all_df.groupby(by="customer_city").agg({
    "order_id": "nunique",
    "payment_value": "sum",
    "product_name": pd.Series.mode
}).nlargest(10, "order_id").reset_index()
bystate_df.rename(columns={
    "order_id": "customer_count"
}, inplace=True)

# Create a horizontal bar plot using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))

# Define bar colors, with the first being different as in your Plotly chart
colors = ["#72BCD4"] + ["#D3D3D3"] * (len(bystate_df) - 1)

# Create the horizontal bar chart
ax.barh(bystate_df['customer_city'], bystate_df['customer_count'], color=colors)

# Add text annotations for each bar
for i in range(len(bystate_df)):
    ax.text(bystate_df['customer_count'][i] + 0.5, i, str(bystate_df['customer_count'][i]), va='center', fontsize=8)

# Set title and labels
ax.set_title('Number of Customers by Cities', fontsize=14)
ax.set_xlabel('Customer Count', fontsize=12)
ax.set_ylabel('City', fontsize=12)

# Show the plot
plt.show()
```

## RFM ANALYSIS

```python
all_df = pd.read_csv("dashboard/clean_data.csv")

datetime_columns = ["order_approved_at","order_delivered_carrier_date",'order_delivered_customer_date','order_estimated_delivery_date' ]
 
for column in datetime_columns:
  all_df[column] = pd.to_datetime(all_df[column])
  all_df[column] = all_df[column].dt.date
  all_df[column] = pd.to_datetime(all_df[column])
all_df.info()
```

```python
df_recency = all_df.groupby(by='customer_id',as_index=False)['order_approved_at'].max()
df_recency.columns = ['CustomerId', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(
    lambda x: (recent_date - x).days)
df_recency.head()
```

```python
frequency_df = all_df.drop_duplicates().groupby(
    by=['customer_id'], as_index=False)['order_approved_at'].count()
frequency_df.columns = ['CustomerId', 'Frequency']
frequency_df.head()
```

```python
monetary_df = all_df.groupby(by='customer_id', as_index=False)['payment_value'].sum()
monetary_df.columns = ['CustomerId', 'Monetary']
monetary_df.head()
```

```python
rf_df = df_recency.merge(frequency_df, on='CustomerId')
rfm_df = rf_df.merge(monetary_df, on='CustomerId').drop(
    columns='LastPurchaseDate')
rfm_df.head()
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)
 
# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['F_rank']/rfm_df['M_rank'].max())*100
 
rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)

rfm_df.head()
```

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.28 * \
    rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['CustomerId', 'RFM_Score']].head(7)
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score'] >
                                      4.5, "Top Customers",
                                      (np.where(
                                        rfm_df['RFM_Score'] > 4,
                                        "High value Customer",
                                        (np.where(
    rfm_df['RFM_Score'] > 3,
                             "Medium Value Customer",
                             np.where(rfm_df['RFM_Score'] > 1.6,
                            'Low Value Customers', 'Lost Customers'))))))
rfm_df[['CustomerId', 'RFM_Score', 'Customer_segment']].head(20)
```

```python
# Count the occurrences of each segment
segment_counts = rfm_df['Customer_segment'].value_counts()

# Define a pastel color palette
color_palette = ['#B7D7D8', '#F3B1B3', '#B5B8C3', '#F9D9B4']

# Create the pie chart
fig, ax = plt.subplots(figsize=(7, 5))

# Create the pie chart with labels, percentage, and colors
wedges, texts, autotexts = ax.pie(
    segment_counts.values, 
    labels=segment_counts.index, 
    autopct='%1.1f%%',  # Display percentages
    colors=color_palette,
    startangle=50,  # Start pie chart from the top
    textprops=dict(color="black", fontsize=7),  # Text font size and color
    wedgeprops=dict(width=0.3, edgecolor='black')  # Adjust wedge width and border
)

# Set the title
ax.set_title('Customer Segmentation', fontsize=16)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Show the plot
plt.show()
```
